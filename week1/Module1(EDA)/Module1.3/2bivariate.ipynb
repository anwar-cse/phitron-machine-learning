sample data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = {
    "Study_Hour": [1, 2, 3, 4, 5, 6, 7, 8],
    
    "Marks": [40, 45, 50, 60, 65, 70, 85, 95],
    
    "Salary": [20000, 22000, 25000, 27000,
               30000, 35000, 40000, 50000],
    
    "Experience": [1, 1, 2, 2, 3, 4, 5, 6],
    
    "Gender": ["Male", "Female", "Male", "Female",
               "Male", "Male", "Female", "Female"],
    
    "Department": ["CSE", "EEE", "CSE", "BBA",
                   "EEE", "CSE", "BBA", "CSE"],
    
    "Result": ["Pass", "Pass", "Pass", "Pass",
               "Pass", "Pass", "Fail", "Pass"]
}
df = pd.DataFrame(data)
print(df)
```

numerical to numerical
scatter plot

```python
plt.figure(figsize=(6,4))

sns.scatterplot(
    x=df["Study_Hour"],
    y=df["Marks"]
)

plt.title("Study Hour vs Marks")

plt.xlabel("Study Hour")

plt.ylabel("Marks")

plt.show()
```

correlation

```python
correlation = df["Study_Hour"].corr(df["Marks"])

print("\nCORRELATION")

print("Correlation:", correlation)
```

Correlation Interpretation

```python
if correlation > 0:
    print("Positive Correlation")

elif correlation < 0:
    print("Negative Correlation")

else:
    print("No Correlation")
```

Correlation Matrix

```python
corr_matrix = df.corr(numeric_only=True)

print(corr_matrix)
```

heatmap

```python
plt.figure(figsize=(7,5))

sns.heatmap(
    corr_matrix,
    annot=True
)

plt.title("Correlation Heatmap")
plt.show()
```

numerical vs categorical
barplot

```python
plt.figure(figsize=(6,4))

sns.barplot(
    x=df["Gender"],
    y=df["Salary"]
)

plt.title("Average Salary by Gender")

plt.show()
```

Boxplot for Category Comparison

```python
plt.figure(figsize=(6,4))

sns.boxplot(
    x=df["Gender"],
    y=df["Salary"]
)

plt.title("Salary Distribution by Gender")

plt.show()
```

categorical vs categorical
crosstab

```python
cross_tab = pd.crosstab(
    df["Gender"],
    df["Result"]
)

print(cross_tab)
```

Groupby Analysis

```python
group_mean = df.groupby("Gender")["Salary"].mean()

print(group_mean)
```

multiple aggregation

```python
agg_result = df.groupby("Department")["Salary"].agg(
    ["mean", "max", "min"]
)

print(agg_result)
```

pairplot

```python
sns.pairplot(df)

plt.show()
```
