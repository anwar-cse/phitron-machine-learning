sample dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

data = {
    "Age": [18, 19, 20, 21, 22, 23, 24, 25, 26, 60],

    "Salary": [20000, 22000, 25000, 27000, 30000,
               32000, 35000, 40000, np.nan, 200000],

    "Gender": ["Male", "Female", "Male", "Male",
               "Female", "Male", "Female",
               "Male", "Male", "Female"]
}

df = pd.DataFrame(data)
print(df)
```

datatype,missing value check, descriptive statistics, value counts(categorical analysis)

```python
print(df.dtypes)
print(df.isnull())
print("missing vulue \n",df.isnull().sum())
print("vulue counts \n",df["Gender"].value_counts())
```

histogram

```python
plt.figure(figsize=(6,4))

sns.histplot(df["Age"], kde=True, bins=5)

plt.title("Histogram of Age")

plt.xlabel("Age")

plt.ylabel("Frequency")

plt.show()
```

KDE Plot

```python
plt.figure(figsize=(6,4))

sns.kdeplot(df["Age"], fill=True)

plt.title("KDE Plot of Age")

plt.xlabel("Age")
plt.show()
```

Boxplot

```python
plt.figure(figsize=(6,4))
plt.boxplot(df["Salary"])
plt.title("Boxplot of salary")
plt.show()
```

countplot

```python
plt.figure(figsize=(7,5))
sns.countplot(x=df["Gender"])
plt.title("Gender countplot")
plt.show()
```

pie chart

```python
df["Gender"].value_counts().plot.pie(
    autopct="%1.1f%%"
)

plt.title("Gender Percentage")

plt.ylabel("")

plt.show()
```
