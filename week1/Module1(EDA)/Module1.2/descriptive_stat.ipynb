sample dataset

```python
import pandas as pd
data = {
    "Name": ["Anwar", "Rahim", "Karim", "Sakib", "Nayeem",
             "Jony", "Mim", "Rafi", "Tania", "Lima"],

    "Marks": [65, 70, 72, 55, 90,
              55, 60, 75, 85, 500]
}
df=pd.DataFrame(data)
print(df)
```

statistics: mean, median,mode, variance, standard deviation, max and min, range,percentile, 

```python
print("Mean value",df["Marks"].mean())
print("median value",df["Marks"].median())
print("mode value",df["Marks"].mode())
print("max frequency",(df["Marks"]==df["Marks"].mode()[0]).sum())
print("variance",df["Marks"].var())
print("standard deviation",df["Marks"].std())
print("max value and min value",df["Marks"].max(),df["Marks"].min())
print("range",df["Marks"].max()-df["Marks"].min())
print("90th percentile",df["Marks"].quantile(0.90))
```

quartiles,  IQR, describe understanding, 

```python
print("quartile \n",df["Marks"].quantile([0.25,0.50,0.75]))
print("IQR",df["Marks"].quantile(0.25)-df["Marks"].quantile(.75))
print("describe \n",df["Marks"].describe())
print("distribution analysis")
if df["Marks"].mean() >df["Marks"].median():
    print("Right Skewed Distribution")
elif df["Marks"].mean()<df["Marks"].median():
    print("Left Skewed Distribution")
else:
    print("Symmetric Distribution")
```

outlier detection using IQR

```python
Q1 = df["Marks"].quantile(0.25)
Q3 = df["Marks"].quantile(0.75)

IQR = Q3 - Q1
print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)

# Outliers detect
outliers = df[
    (df["Marks"] < lower_bound) |
    (df["Marks"] > upper_bound)
]

print("\nOutliers:")
print(outliers)
```
